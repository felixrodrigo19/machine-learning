```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('../input/airbnbopendata/Airbnb_Open_Data.csv')
df.head()
```

```python
columns = ['id',
           'host id',
           'host_identity_verified',
           'neighbourhood group',
           'neighbourhood',
           'country', 
           'service fee',
           'minimum nights',
           'number of reviews',
           'reviews per month',
           'review rate number',
           'calculated host listings count',
           'availability 365',
           'house_rules']
```

```python
df = df[columns]
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.sum()
```

```python
df['review rate number'].unique()
```

```python
plt.plot(df['review rate number'].value_counts())
plt.ylabel('review rate number')
plt.show()
```

```python
df['review rate number'].value_counts().plot().bar()
```

```python
fig, ax = plt.subplots()
ax.scatter(df['review rate number'], df['service fee'].to_string)

ax.set_xlabel('review rate number', fontsize=15)
ax.set_ylabel('review rate number', fontsize=15)

ax.grid(True)
fig.tight_layout()

plt.show()
```
